```python
def tab(url):
    browser.get(url)
    html = browser.page_source
    soup = BeautifulSoup(html,"html.parser")
    title = []
    add  = []
    rating= []
    price = []
    box = soup.find_all("div", class_="hotelCardListing__descriptionWrapper")
    for i in box:
        t = i.find_all("h3", class_="listingHotelDescription__hotelName d-textEllipsis")
        if(len(t)!=0):
            title.append(t[0].text)
        else:
            title.append("Not Found")
        l = i.find_all("span", class_="u-line--clamp-2")
        if(len(l)!=0):
            add.append(l[0].text)
        else:
            add.append("Not Found")
        r = i.find_all("div", class_="hotelRating__wrapper")
        if(len(r)!=0):
            rating.append(r[0].text.split(" ")[0].strip())
        else:
            rating.append("Not Found")
        p = i.find_all("span", class_="listingPrice__finalPrice")
        if(len(p)!=0):
            price.append(p[0].text)
        else:
            price.append("Not Found")
    table = pd.DataFrame({"Title":title, "Address":add,"Ratings":rating,"Price":price})
    return table

tab("https://www.oyorooms.com/hotels-in-coimbatore/?page=4")
```

```python
base_url ="https://www.oyorooms.com/hotels-in-coimbatore/?page="
```

```python
table = {}
for i in range(1,6):
    url = base_url+str(i)
    table[i] = tab(url)
```
